# pie chart

#### when we want to show %wise distribution of a data

```python
teams = ['MI','CSK','KKR','RCB']
win = [20,18,10,5]
```

#### pie chart is used for Univariate analysis

```python
plt.pie(win)
plt.show()
```

```python
#mention the labels of pie
plt.pie(win,labels=teams)
plt.show()
```

```python
#change color
plt.pie(win,labels=teams,colors=['b','y','m','r'])
plt.show()
```

```python
#percentage wise distribution
plt.pie(win,labels=teams,colors=['b','y','m','r'],autopct='%.2f%%')
plt.show()
```

```python
#use explode to seperate the pie
plt.pie(win,labels=teams,colors=['b','y','m','r'],autopct='%.2f%%',explode=[0.2,0.1,0,0])
plt.show()
```
